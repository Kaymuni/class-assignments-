```python
import sympy as sp

def solve_second_order_ode():
    # Define symbols
    x = sp.Symbol('x')
    y = sp.Function('y')

    print("--- Second Order ODE Solver (ay'' + by' + cy = f(x)) ---")

    try:
        # Input coefficients and the function f(x)
        a = float(input("Enter coefficient a: "))
        b = float(input("Enter coefficient b: "))
        c = float(input("Enter coefficient c: "))
        f_input = input("Enter f(x) (e.g., 'exp(x)', 'sin(x)', or '0'): ")
        f = sp.sympify(f_input)

        # Define the ODE: a*y''(x) + b*y'(x) + c*y(x) = f(x)
        ode = sp.Eq(a * y(x).diff(x, x) + b * y(x).diff(x) + c * y(x), f)

        # 1. Find the Homogeneous Solution (y_h)
        # Solve the ODE with f(x) = 0
        ode_h = sp.Eq(a * y(x).diff(x, x) + b * y(x).diff(x) + c * y(x), 0)
        sol_h = sp.dsolve(ode_h, y(x))
        y_h = sol_h.rhs

        # 2. Find the Particular Solution (y_p)
        # Full solution minus the homogeneous part
        full_sol_gen = sp.dsolve(ode, y(x))
        # To isolate y_p, we set the arbitrary constants C1 and C2 to 0
        y_p = full_sol_gen.rhs.subs({sp.Symbol('C1'): 0, sp.Symbol('C2'): 0})

        print(f"\nHomogeneous Solution (y_h): {y_h}")
        print(f"Particular Solution (y_p): {y_p}")
        print(f"General Solution: y(x) = {full_sol_gen.rhs}")

        # 3. Handle Auxiliary Conditions
        print("\n--- Input Auxiliary Conditions ---")
        x0 = float(input("Enter x value for conditions (x0): "))
        y0 = float(input(f"Enter y({x0}): "))
        dy0 = float(input(f"Enter y'({x0}): "))

        # Solve for constants C1 and C2
        # y(x0) = y0
        cond1 = sp.Eq(full_sol_gen.rhs.subs(x, x0), y0)
        # y'(x0) = dy0
        cond2 = sp.Eq(full_sol_gen.rhs.diff(x).subs(x, x0), dy0)

        constants = sp.solve((cond1, cond2), (sp.Symbol('C1'), sp.Symbol('C2')))

        # Final Solution
        final_sol = full_sol_gen.rhs.subs(constants)

        print("\n--- Results ---")
        print(f"Constants found: {constants}")
        print(f"Final Solution: y(x) = {final_sol}")

        return y_h, y_p, final_sol

    except Exception as e:
        print(f"An error occurred: {e}")

# Run the function
if __name__ == "__main__":
    solve_second_order_ode()
```
